```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("brain_size.csv",sep=";")
df.head()
```

```python
# Compute basic statistics - shape, no. of columns, mean
print("Shape of the dataset is: ", df.shape)
print("Number of Columns in the dataset: ", len(df.axes[1]))
print("Number of Rows in the dataset: ", len(df.axes[0]))
print("Mean of FSIQ col in the dataset: ", df["FSIQ"].mean())
```

```python
# Categorically Split the Data based on Gender Column
SD = df["Gender"].unique()
for gender in SD:
    print(df[df["Gender"]==gender])
```
