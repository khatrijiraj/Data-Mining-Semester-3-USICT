```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df1 = df.copy()
df1.isnull().sum()
```

```python
# Remove rows/Attributes
df1.dropna(inplace = True)
```

```python
Age_mean = df["Age"].mean().round()
df["Age"].fillna(value = Age_mean, inplace = True)
```

```python
Embarked_mode = df["Embarked"].mode()[0]
#Embarked_mode
df["Embarked"].fillna(value = Embarked_mode, inplace = True)
```

```python
Cabin_mode = df["Cabin"].mode()[0][0:3]
#Cabin_mode
df["Cabin"].fillna(value = Cabin_mode, inplace = True)
```

```python
df1.isnull().sum()
```

```python
#Transformation & Discretization
```

```python
df['Age_Range'] = pd.cut(df['Age'], bins=[0,9,18,60,100], labels=["Child","Teenager","Adult", "Aged"])
df.head()
```

```python
df['Age_minmax'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df
```
