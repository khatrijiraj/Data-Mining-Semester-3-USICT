```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix,accuracy_score
```

### Linear Regression Example ###

```python
# Generate simple synthetic data for Linear Regression (single feature)
np.random.seed(0)
X_lin = np.random.rand(50, 1) * 10  # Random values between 0 and 10
y_lin = 2.5 * X_lin.squeeze() + 3 + np.random.randn(50) * 2  # y = 2.5 * x + 3 + noise
```

```python
# Train-test split
X_train_lin, X_test_lin, y_train_lin, y_test_lin = train_test_split(X_lin, y_lin, test_size=0.2, random_state=0)
```

```python
# Initialize and train Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train_lin, y_train_lin)

# Predict on test set
y_pred_lin = linear_model.predict(X_test_lin)

# Plot results for Linear Regression
plt.figure(figsize=(8, 5))
plt.scatter(X_test_lin, y_test_lin, color="blue", label="Actual")
plt.plot(X_test_lin, y_pred_lin, color="red", linewidth=2, label="Predicted")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression - Actual vs Predicted")
plt.legend()
plt.show()
```

```python
# Performance metrics for Linear Regression
print("Linear Regression Performance:")
print("Mean Squared Error (MSE):", mean_squared_error(y_test_lin, y_pred_lin))
print("R-squared Score (R2):", r2_score(y_test_lin, y_pred_lin))
# print(accuracy_score(y_test_lin,y_pred_lin))
```

### Logistic Regression Example ###

```python
# Generate simple synthetic data for Logistic Regression (binary classification)
np.random.seed(0)
X_log = np.vstack((np.random.randn(25, 2) + [2, 2], np.random.randn(25, 2) - [2, 2]))  # Two clusters
y_log = np.hstack((np.ones(25), np.zeros(25)))  # Labels for clusters

# Train-test split
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X_log, y_log, test_size=0.2, random_state=0)

# Initialize and train Logistic Regression model
logistic_model = LogisticRegression()
logistic_model.fit(X_train_log, y_train_log)

# Predict on test set
y_pred_log = logistic_model.predict(X_test_log)

# Plot results for Logistic Regression with decision boundary
plt.figure(figsize=(8, 5))
plt.scatter(X_log[y_log == 0][:, 0], X_log[y_log == 0][:, 1], color="blue", label="Class 0")
plt.scatter(X_log[y_log == 1][:, 0], X_log[y_log == 1][:, 1], color="red", label="Class 1")

# Decision boundary
x_values = np.linspace(X_log[:, 0].min(), X_log[:, 0].max(), 100)
y_values = -(logistic_model.coef_[0][0] * x_values + logistic_model.intercept_[0]) / logistic_model.coef_[0][1]
plt.plot(x_values, y_values, color="green", linestyle="--", label="Decision Boundary")

plt.xlabel("X)
plt.ylabel("Y")
plt.title("Logistic Regression - Decision Boundary")
plt.legend()
plt.show() 

# Performance metrics for Logistic Regression
print("Logistic Regression Performance:")
# print("Accuracy:", accuracy_score(y_test_log, y_pred_log))
print("\nConfusion Matrix:\n", confusion_matrix(y_test_log, y_pred_log))
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X1 = np.random.normal(2, 1, (50, 2))  # Class 1 points centered at (2,2)
X2 = np.random.normal(4, 1, (50, 2))  # Class 2 points centered at (4,4)

X = np.vstack((X1, X2))  # Combine data
y = np.hstack((np.zeros(50), np.ones(50)))  # Labels: 0 for Class 1, 1 for Class 2

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize and train Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on test set
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Visualize data points
plt.figure(figsize=(10, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='bwr', edgecolor='k', label='Train Data')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='cool', edgecolor='k', marker='x', label='Test Data')

# Plot decision boundary
x_values = np.linspace(min(X[:, 0]) - 1, max(X[:, 0]) + 1, 100)
y_values = -(model.coef_[0][0] * x_values + model.intercept_[0]) / model.coef_[0][1]
plt.plot(x_values, y_values, color='black', linestyle='--', label='Decision Boundary')

# Show predicted points with labels
for i in range(len(y_pred)):
    color = 'red' if y_pred[i] == 1 else 'blue'
    plt.scatter(X_test[i, 0], X_test[i, 1], color=color, marker='o', s=80, alpha=0.4)

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Logistic Regression - Accuracy: {accuracy:.2f}')
plt.legend()
plt.show()
```
